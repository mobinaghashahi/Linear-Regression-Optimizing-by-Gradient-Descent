```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generateData():

    
    num_points = 100
    start = -5
    end = 5
    X = np.linspace(start, end, num_points)
    
    Y=10*X+4
    
    mean = 0  
    std_dev = 5  
    noise = np.random.normal(mean, std_dev, num_points)
    Y=Y+noise
    return X,Y
```

```python
def costFunction(x,y,w):
    return np.sum((np.dot(x,w.T) - y)**2) / len(x)
```

```python
def drawLine(x,w):
    return x*w[0][1]+w[0][0]
```

```python
def gradietnDesent(x,y,w,iteration,step):
    cost_history = []
    w_history = []
    for i in range(iteration):
        print(f"w0={w[0][0]} w1={w[0][1]}")
        print(f"cost={np.sum((y-np.dot(x,w.T))*x)}")
        cost_history.append(costFunction(x, y, w))
        w0_history.append(w[0][0].copy())
        w1_history.append(w[0][1].copy())
        w[0][0]-=((2*step/len(y))*(np.sum((np.dot(x,w.T)-y))))
        w[0][1]-=((2*step/len(y))*(np.sum((np.dot(x,w.T)-y)*x)))
    return w,cost_history,w_history
```

```python
#generate Data
x,y=generateData()

#add bias to the data
ones_column = np.ones((x.shape[0], 1))  
x = x.reshape(-1, 1)  # تغییر شکل به (100, 1)
y = y.reshape(-1, 1)
X_with_bias = np.hstack((ones_column,x)) 


#initial w
w=np.array([[20, 40]])


w,cost_history,w_history=gradietnDesent(X_with_bias,y,w,100,0.01)
```

```python

print(len(w_history[0][0]))
plt.scatter(x, y, color='blue', label='Actual Data')
plt.plot(x, drawLine(x, w), color='red', label='Prediction (Closed Form)')
plt.title(f"Polynomial Regression (Degree {1})")
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.legend()
plt.show()
```

```python
# داده‌های نمونه
X = w[0][1]
Y = w[0][1]

Z = np.sin(np.sqrt(X + Y**2))

# ایجاد نمودار
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# رسم سطح
ax.plot_surface(X, Y, Z, cmap='viridis')

# تنظیم برچسب‌ها
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
